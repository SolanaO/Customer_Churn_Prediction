## <font color='blue'>Load, Clean and Save Data</font>

```python
def load_data(file_path):
    """
    Loads the raw dataset in Spark.
    
    INPUT:
            (str) - path for datafile
    OUTPUT:
            (PySpark dataframe) - dataframe of raw data
    
    """
    
    print("Loading the dataset ...")
    df = spark.read.json(file_path)
    print("Dataset is loaded...")
    
    return df
```

```python
def clean_data(df):
    
    """
    Performs basic cleaning operations on the raw data:
        - removes entries with missing userId
        - parse timestamps columns
        - drop unnecessary columns
    
    INPUT:
            (PySpark dataframe) - dataframe of raw data
    OUTPUT:
            (PySpark dataframe) - dataframe of cleaned data
    """
    
    # print message to indicate the start of the process
    print("Cleaning the data ...")
    
    # print a count of rows before cleaning
    initial_records = df.count()
    print("Dataset has {} rows initially.".format(initial_records))
    
    # filter out all the records without userId
    df = df.where(df.userId != "")
    
    # parse the timestamp columns: registration and ts
    df = df.withColumn("date_ts", from_unixtime(col("ts")/1000.0))
    df = df.withColumn("date_reg", from_unixtime(col("registration")/1000.0))

    # drop several columns 
    cols_to_drop = ("firstName", "lastName", "location",
                    "artist", "song", "length",
                    "userAgent",
                    "registration", "ts")
    df = df.drop(*cols_to_drop)

    
    # print end of process message
    print("Finished cleaning the data ...")
    
    # print a count of rows after cleaning
    removed_rows = initial_records - df.count()
    print("Cleaned dataset has {} rows, {} rows were removed". \
        format(df.count(), initial_records - df.count()))

    return df
```

```python
# specify datafile information
file_path = "data/mini_sparkify_event_data.json"
```

```python
# load the dataset
df1 = load_data(file_path)
```

```python
# clean the data
df1_clean = clean_data(df1)
```

```python
# check for null values to confirm cleaned data
df1_clean.select([count(when(isnan(c) | col(c).isNull(), c)) \
           .alias(c) for c in df1_clean.columns]) \
           .show(vertical=True)
```

```python
# save the cleaned dataset to json 
clean_data_path = "data/mini_sparkify_clean_data.json"
df1_clean.write.json(clean_data_path)
```
